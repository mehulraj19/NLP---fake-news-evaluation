# Importing Packages 

```python
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

# Importing Dataset

```python
df = pd.read_csv('news.csv')
```

```python
df = df[['title', 'text', 'label']]
```

```python
df = df.dropna()
```

```python
df.shape
```

# NLP Data Pre-Processing

```python
list_of_stopwords = list(stopwords.words('english'))
```

```python
clean_text_main = []
for i in range(df.shape[0]):
    text_message = df['title'].values[i]
    tokenized_word = word_tokenize(text_message)
    clean_text = ''
    for word in tokenized_word:
        if not word in list_of_stopwords:
            clean_text += word + ' '
    clean_text_main.append(clean_text)
len(clean_text_main)
```

```python
df['clean_text'] = clean_text_main
```

# Data Split for model trainning

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
sparse_matrix = cv.fit_transform(df['clean_text'])
```

```python
le = LabelEncoder()
df['label_main'] = le.fit_transform(df['label'])
```

# Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(sparse_matrix, df['label_main'], random_state=0, train_size=0.8) 
```

# Model Trainning

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

# Model Evaluation

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```
